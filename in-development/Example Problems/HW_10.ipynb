## Problem 1: x-y Diagram for Methanol/Water

Methanol is a by-product of common pharmaceutical fermentation processes. For the purpose of designing a binary distillation column to strip the methanol from the fermentation broth, construct an x-y diagram using Raoult's law and Antoine's equations for an operating pressure of 97.99 kPa. Plot the experimental data from http://www.ddbst.com/en/EED/VLE/VLE%20Methanol%3BWater.php on the same plot.


## Problem 2: Binary Distillation

A particular pharmaceutical fermentation produces a methanol/water mixture that is 8 mol% methanol. Design a distillation column to strip 90% of the methanol out of the fermentation broth, and produce a methanol product that is at least 60 mol% methanol.

## Problem 3: Cardiac Energy

Under moderate exercise, an athlete is observed to have a heart beat of 160 bpm, a stroke volume of 90 ml, and a systolic pressure 190 mmHg.  Assuming that blood has a density of 1.025 g/ml, how much power is generated by the heart in watts?  In horsepower?

## Problem 4: Cooling Packs

Design a 200g cooling pack using sodium carbonate and water that will reach a temperature of (2 deg C) starting at 25 deg C,. You can assume the sodium carbonate/water solution as a specific heat of 4.2 kJ/kg.

## Problem 5: Steam Catapult

A 30,000 kg aircraft is to be launched from the deck of an aircraft carrier using a steam catapult. The 100 meter long catapult must accelerate the aircraft to a take-off speed of 80 meters/second. The moving portions of the catapult add 3,000 kg to the accelerate mass. To a rough approximation, all of the energy for the launch comes from a steam catapault. 

Assuming a constant rate of acceleration, how much power is generated by the catapult in kW? In horsepower?

The catapult is initially charged steam at 40 bar (absolute) and 350 deg C. Following launch, the steam pressure inside the catapult is measured to be 5 bar (absolute). Assuming that 20% of the steam will condense at the exit conditions, how much steam will be required?
